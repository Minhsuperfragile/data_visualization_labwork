## Load data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homework_sampledata1.txt')
df.head()
```

## Exercise 1
Print value of the first day of each month

```python
cols = df.columns[1:]
for i in cols:
    print(f"2008-{i if len(i) > 1 else f'0{i}'}-01 00:00:00 {df[i][0]}")
```

## Exercise 2
Get rid of missing values

```python
#I want to replace with None so that matplotlib will plot a discontinuous line
df.replace(-99, None, inplace=True)
df.tail()
```

## Exercise 3:
Plot the daily temperature from 01 Jan 2008 to 31 Dec 2008. Save the file as 'Figure1_1.png'

```python
data_value = []
for col in cols:
    if col == '2':
        values = df[col][:-2]
    elif col in ['4','6','9','11']:
        values = df[col][:-1]
    else:
        values = df[col]
        
    for value in values:
        data_value.append(value)
```

```python
time_range = range(len(data_value))

plt.figure(figsize=(15,5))
plt.plot(time_range, data_value, linestyle='solid')
plt.grid()
plt.ylim(5, 35)
plt.xlim(0, 365)
plt.title("Temperature in 2008")
plt.xlabel("Time")
plt.ylabel("Temp(oC)")
plt.xticks([0, 31, 58, 89, 119, 150, 180, 211, 242, 272, 303, 333], ["01\n Jan 2008", "01\n Feb", "01\n Mar", "01\n Apr", "01\n May", "01\n Jun", "01\n Jul", "01\n Aug", "01\n Sep", "01\n Oct", "01\n Nov", "01\n Dec"])
plt.legend(['Temperature'])
plt.savefig('Figure1_1.png')
plt.show()
```

## Exercise 4
Calculate the monthly mean in 2008. Please note that the days in different months are different. Save the file as 'Figure1_2.png'.

```python
means = []

for col in cols:
    means.append(df[col].mean())

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov","Dec"]

plt.figure(figsize=(15,5))
plt.ylim(10, 28)
plt.plot(months, means)
plt.plot(means, 'bo')
plt.legend(['Temperature'])
plt.xlabel('Month')
plt.ylabel('Temperature(oC)')
plt.title('Average temperature of each month in 2008')
plt.savefig('Figure1_2.png')
plt.grid()
```
